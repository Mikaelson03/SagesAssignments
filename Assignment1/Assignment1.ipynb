```python
# ENTER NUMBER
number = input("Please enter a number:\n")

# CHECK IF NUMBER IS POSITIVE, NEGATIVE, OR ZERO
if "-" in number:
    print("The number is negative.")
elif number == "0":
    print("The number is zero.")
else:
    print("The number is positive.")

# CHECK IF NUMBER IS EVEN OR ODD
if int(number) % 2 == 0:
    print("The number is even.")
else:
    print("The number is odd.")

# CHECK IF THE NUMBER IS BETWEEN 10 AND 100
if 10 < int(number) < 100:
    print("The number is between 10 and 100.")
else:
    print("The number is not between 10 and 100.")
```

```python
# PART A
fruits = ["apple", "banana", "agbalumo", "pineapple", "strawberry"]
for index, fruit in enumerate(fruits):
    print(f"{index+1}. {fruit}")

# PART B
password = input("Please enter a password:\n")
tries = 0
while password not in fruits:
    tries += 1
    if tries >= 3:
        print("Too many attempts. Access denied.")
        break
    print("Invalid password.")
    password = input("Please enter a password:\n")

if password in fruits:
        print("Access granted.")
```

```python
import pandas as pd
import numpy as np

# PART A
scores = [85, 66, 78, 48, 33, 48]
scores_array = np.array(scores)
print(scores_array.shape)
print(scores_array.dtype)
print("mean:", f"{np.mean(scores_array):.3f}")
print("min:", np.min(scores_array))
print("max:", np.max(scores_array))
print("std:", f"{np.std(scores_array):.3f}")

scores_grt_60 = scores_array[scores_array > 60]
print("Scores greater than 60:", scores_grt_60)

# PART B

students_data = {
    "name": ["Joseph", "Freeman", "Elijah", "Klaus", "Rebekah"],
    "age": [25, 30, 28, 32, 27],
    "score": [85, 66, 78, 48, 33],
    "grade": ["A", "B", "B", "D", "F"]
}

students_df = pd.DataFrame(students_data)

# FIRST 3 ROWS
print(students_df.head(3))

# SCORE ABOVE 50
print(students_df[students_df["score"] > 50])

# SORT BY SCORE IN DESCENDING ORDER
print(students_df.sort_values(by="score", ascending=False))

# ADD "PASSED" COLUMN TO DF
students_df["passed"] = np.where(students_df["score"] >= 50, True, False)
print(students_df)

students_df.to_csv("class_results.csv", index=False)

s_df = pd.read_csv("class_results.csv")
print(s_df)

```
